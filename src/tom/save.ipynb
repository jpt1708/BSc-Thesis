```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
min_cpu = 50
max_cpu = 100

traffic = np.random.uniform(min_cpu, max_cpu, 10000)
req = np.arange(0, 10000)
y = traffic * (((np.sin((req * 2 * np.pi) / 10000) + 1) / 2) * 0.9 + 0.1)

plt.figure(figsize=(8, 8))
plt.scatter(req, y)

print(y)
print(min(y), max(y))
print((y < min_cpu).sum())
```
